## Set missing values for incomplete data (i.e. include 12 but with missing values)

### - Hand0

```python
hand0_8 = dd_0_8_df['ID'].unique()

hand0_tmp = dd_0_8_df.copy()

for cid in hand0_8:
    # Get the data for the current id
    aux = hand0_tmp[hand0_tmp['ID'] == cid]
    # Get the maximum Follow-up registered
    fu = max(aux['FU'])
    # Get the last row so we can add extra rows
    last_row = aux[aux['FU'] == fu]

    # Repeat the process until we complete 12
    last_month = max(aux['MONTH'])
    for add_fu in range(fu + 1, 13):
        to_append = last_row.copy()
        last_month += 6
        index = to_append.index[0]
        to_append.set_value(index, 'FU', add_fu)
        to_append.set_value(index, 'MONTH', last_month)
        to_append.set_value(index, 'AREA', np.NaN)
        to_append.set_value(index, 'ANGLE', np.NaN)
#         print(to_append)
        hand0_tmp = hand0_tmp.append(to_append, ignore_index = True)

hand0_tmp[hand0_tmp['ID'] == 46]
```

```python
hand1_8 = dd_1_8_df['ID'].unique()

hand1_tmp = dd_1_8_df.copy()

for cid in hand1_8:
    # Get the data for the current id
    aux = hand1_tmp[hand1_tmp['ID'] == cid]
    # Get the maximum Follow-up registered
    fu = max(aux['FU'])
    # Get the last row so we can add extra rows
    last_row = aux[aux['FU'] == fu]

    # Repeat the process until we complete 12
    last_month = max(aux['MONTH'])
    for add_fu in range(fu + 1, 13):
        print(last_month)
        to_append = last_row.copy()
        index = to_append.index[0]
        to_append.set_value(index, 'FU', add_fu)
        to_append.set_value(index, 'MONTH', last_month)
        to_append.set_value(index, 'AREA', np.NaN)
        to_append.set_value(index, 'ANGLE', np.NaN)
#         print(to_append)
        hand1_tmp = hand1_tmp.append(to_append, ignore_index = True)
```
